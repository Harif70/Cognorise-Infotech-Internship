# Unemployment in India
# **Context**

All of India's states' unemployment rates are included in this dataset.

Region: Indian states


Date = the day on which the rate of unemployment was noted

Frequency = Monthly Frequency Measurement

Estimated Labour Participation Rate (%) = labour force participation rate by dividing the number of people actively participating in the labour force by the total number of people eligible to participate in the labor force

Percentage of persons employed equals the estimated employed.

The labor force participation rate is calculated by dividing the total number of individuals eligible to join the labor force by the number of individuals actively engaging in the labor force. This yields the estimated labor participation rate (%).

# Problem:
  The dataset facilitates an understanding of the variations in unemployment trends among different states in India during the COVID-19 crisis. It provides valuable information on the effects on the unemployment rate, employment statistics, and labor participation rates in diverse regions of the country.

```python
df2 = unemployment_data.groupby('Region')[[' Estimated Employed']].mean().sort_values(by=' Estimated Employed',ascending= False)
df2.plot(kind='bar')
plt.title('Average millions employed per region')
plt.ylabel('Millions employed')
plt.show()
```

```python
df2 = unemployment_data.groupby('Region')[[' Estimated Unemployment Rate (%)']].mean().sort_values(by=' Estimated Unemployment Rate (%)',ascending= False)
df2.plot(kind='bar')
plt.title('Estimated unemployment percentage per region')
plt.ylabel('Unemployment percentage')
plt.xlabel('Region')
plt.xticks(rotation= 90)
```
